```python
import matplotlib.pyplot as plt
import os
```

```python
entry = r"."
means = []

with open(os.path.join(entry, "snapshot.txt"), "r") as f:
    for line in f.readlines():
        line = line.rstrip()
        if "mean" in line:
            means.append( float(line.split()[3]) )


means[:5]
```

```python
plt.style.use('ggplot')

plt.title("Mean of Score")
plt.plot(range(1000, 1000*(len(means)+1), 1000), means, "b-")

plt.xlabel("episode")
plt.ylabel("score")

plt.show()
```
